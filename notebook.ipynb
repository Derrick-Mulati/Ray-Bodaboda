# Data cleaning
## cheaking for duplicates

```python
# A function to check for duplicates in our datasets
def check_duplicates(df):
    """
    This function checks for and returns any duplicates in a given dataframe.
    """
    duplicates = df[df.duplicated()]
    if duplicates.shape[0] == 0:
        print("No duplicates found in the dataset")
    else:
        print("Duplicates found in the dataset:")
        return duplicates
```

```python
# Calling for the function to check for duplicates
check_duplicates(data)
```

#### Cheaking for missing values

```python
# A function to check for missing values in our dataset
def check_missing_values(data):
    # Count missing values in each column
    missing_values = data.isnull().sum()

    # Convert missing values count to percentage of total rows
    missing_percent = (missing_values / len(data)) * 100

    # Combine the missing values count and percent into a DataFrame
    missing_df = pd.concat([missing_values, missing_percent], axis=1)
    missing_df.columns = ['Missing Values', '% of Total']

    # Return only columns with missing values
    missing_df = missing_df[missing_df['Missing Values'] > 0]

    return missing_df

# Check missing values in each dataset
display(check_missing_values(data))
```

# Feature Engineering
Feature engineering is the process of creating new features from existing data to improve the performance of a machine learning model. It involves selecting, transforming, and creating new variables that capture important patterns, relationships, or domain knowledge in the data.

* Age Group: Create a new feature by categorizing the "Age" variable into groups 

* Brand and Model Category: Extract the brand name from the "Motorcycle Brand and Model" variable and create a new categorical feature representing the brand category (e.g., "Boxer", "Tvs", "Honda"). This can help capture any brand-specific patterns.

* Engine Power: This can provide a more meaningful representation of the engine power for analysis or modeling purposes.

* Loan Facility: Create a binary feature indicating whether a loan facility was facilitated by a financial institution ("Bank", "Micro Finance", "SACCO") or not ("Not Applicable", "A friend or family"). This can help assess the impact of loan facilities on bike ownership and usage patterns.

* Monthly Income: Calculate the monthly income by multiplying the "How much do you make daily?" variable with the average number of working days in a month. This can provide a more standardized measure of income for analysis or modeling.

* Fuel Efficiency: Create a new feature by dividing the "How many liters of fuel do you typically consume per day?" variable by the "How Many Kilometers do you cover in a day?" variable. This can represent the fuel efficiency of the motorcycles and can be useful for understanding fuel consumption patterns.

* Customer Satisfaction: Calculate an average customer satisfaction score by dividing the "Averagely how many customers do you serve in a day?" variable by the "How many liters of fuel do you typically consume per day?" variable. This can provide an indication of customer satisfaction per unit of fuel consumption.

* Insurance Cost: Calculate the insurance cost per month by dividing the "If you answered yes above, how much do you spend on insurance?" variable by the number of months in a year. This can provide a standardized measure of insurance cost for analysis or modeling purposes.
